```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot

## sns.scatterplot(): Visualizes the relationship between two numeric variables,
## helping identify correlations, clusters, and patterns.

## sns.boxplot(): Displays the distribution of a numeric variable across different categories,
## showing the median, quartiles, and outliers for comparison.

## sns.histplot(): Shows the distribution of a single numeric variable, helping understand
## frequency distribution, skewness, and modality.

## sns.barplot(): Displays the mean or other central tendency of a numeric variable for
## different categories, useful for comparing categorical data.

## sns.heatmap(): Visualizes data in a matrix form with colors representing values,
## useful for displaying correlations, matrices, and data distribution across two dimensions.
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.
# Note:  timepoint, signal, event, and region are columns in the fmri dataset

data = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=data)
plt.show()
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
# 'pclass', y = 'age' and y = 'fare'.

# Note:  pclass, age, and fare are columns in the titanic dataset.

data = sns.load_dataset("titanic")
sns.boxplot(x="pclass", y="age", data=data)
plt.show()
sns.boxplot(x="pclass", y="fare", data=data)
plt.show()
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.


data = sns.load_dataset("diamonds")
sns.histplot(data=data, x="price", hue="cut")
plt.show()
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.

data = sns.load_dataset("iris")
sns.pairplot(data, hue="species")
plt.show()
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

data = sns.load_dataset("flights")
plt.figure(figsize=(15, 6))
flights_pivot = data.pivot(index="month", columns="year", values="passengers")
sns.heatmap(flights_pivot, annot=True, cmap="YlGnBu")
plt.show()
```
